```python
from sklearn.datasets import load_iris
```

```python
data = load_iris()
```

```python
X = data.data[:,:-2]
```

```python
print(X)
```

```python
import matplotlib.pyplot as plt
```

```python
plt.scatter(X[:,0] , X[:,1])
```

```python
from sklearn.cluster import KMeans
```

```python
clu = KMeans(n_clusters = 2)
```

```python
clu.fit(X)
```

```python
pred = clu.predict(X)
pred
```

```python
c_x , c_y = clu.cluster_centers_[:,0] , clu.cluster_centers_[:,1]
```

```python
plt.scatter(X[:,0] , X[:,1])
plt.scatter(c_x,c_y , color = 'red')
```

```python
colors = {1: 'blue' , 0: 'green'}
colors
```

```python
for(x,y) , group in zip(X,pred):
    plt.scatter(x,y , c= colors[group])
    plt.scatter(c_x , c_y , color = 'red')
```

```python
X = data.data[:,:-2]
```

```python
errors = list()
for k in range (1,10):
    clu = KMeans(n_clusters = k)
    clu.fit(X)
    preds = clu.predict(X)
    errors.append(clu.inertia_)
```

```python
errors
```

```python
plt.plot(list(zip(range(1,10) , errors )))
```

```python
plt.scatter(range(1,10) , errors)
```

```python
from sklearn.model_selection import GridSearchCV as GS 
```
